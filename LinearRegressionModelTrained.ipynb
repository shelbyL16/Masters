```python
# Load dataset - NESARC Alcohol
data2 = pd.read_table("//cbio//projects//018//data//prs//nesarc//FinalPRSFull.txt", delim_whitespace=True)
np.random.seed(42)

# create the y for the nesarc dataset
y=data2["LIFEAUD5"]
y
```
